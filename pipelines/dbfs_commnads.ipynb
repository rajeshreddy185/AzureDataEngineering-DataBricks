
| Action | Magic (%fs) | Python (dbutils) | CLI (databricks fs) |
|--------|-------------|-----------------|--------------------|
| List   | %fs ls /path | dbutils.fs.ls(""/path"") | ls dbfs:/path |
| Copy   | %fs cp /src /dst | dbutils.fs.cp(""/src"", ""/dst"") | cp dbfs:/src dbfs:/dst |
| Move   | %fs mv /src /dst | dbutils.fs.mv(""/src"", ""/dst"") | mv dbfs:/src dbfs:/dst |
| Make Dir | %fs mkdirs /path | dbutils.fs.mkdirs(""/path"") | (Implicit on copy) |
| Remove | %fs rm /path | dbutils.fs.rm(""/path"") | rm dbfs:/path |
| Help   | %fs help | dbutils.fs.help() | --help |
